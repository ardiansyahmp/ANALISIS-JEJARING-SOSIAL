```python
df=pd.read_csv("WonderfulIndonesia.csv")
df
```

```python
#Menghapus baris yang terdapat missing value 
df = df.loc[:, ['id', 'Username']]
df.dropna()
df.head()
```

```python
graph_temp = nx.from_pandas_edgelist(df,source="id",target="Username")
nx.info(graph_temp)
graph_temp.nodes()
print('Panjang Nodes',len(graph_temp.nodes()))
```

```python
print('Panjang Edges',len(graph_temp.edges()))
```

```python
bet_centrality = nx.betweenness_centrality(graph_temp, normalized = True, 
                                              endpoints = False) #calculate using betweness centrality
  
print(bet_centrality)
```

```python
# Analisis
# Menghitung derajat dari nodes
temp = []
tap = nx.degree(graph_temp)
for i in tap:
    temp.append([i[0],i[1]])

sort_deg = pd.DataFrame(temp)
sort_deg.sort_values(1,axis=0,ascending = False, inplace = True)
sort_deg=sort_deg.rename(columns={0: "Username", 1: "Degree"})
sort_deg.head()
```

```python
# Mengidentifikasi Most Influential User dengan menghitung betweeness centrality
save = []
most_influential = nx.betweenness_centrality(graph_temp)

for w in sorted(most_influential, key=most_influential.get, reverse = True):
    save.append([w, most_influential[w]])
    
df_clos = pd.DataFrame(save)
df_clos = df_clos.rename(columns={0:"Username", 1:"Betweeness Centrality"})
df_clos.head()
```
